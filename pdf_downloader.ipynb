```python
EXCEL_FILE = '/Users/narainyucel/Google Drive/MADS/capstone/legal_nlp/code/DTMB_PRO-Contract-List.xls'
OUTPUT_DIR = '/Users/narainyucel/Google Drive/MADS/capstone/legal_nlp/code/test_data'

with xw.App(visible=False) as app:
    wb = app.books.open(EXCEL_FILE)
    for sheet in wb.sheets:
        wb_new = app.books.add()
        sheet.copy(after=wb_new.sheets[0])
        wb_new.sheets[0].delete()
        if not os.path.isfile(OUTPUT_DIR + f'/{sheet.name}.xlsx'):
            wb_new.save(OUTPUT_DIR + f'/{sheet.name}.xlsx')
        wb_new.close()
```

```python
import io
from xlsx2html import xlsx2html

# must be binary mode
xlsx_file = open('Updated 06-16-2022.xlsx', 'rb') 
out_file = io.StringIO()
xlsx2html(xlsx_file, out_file, locale='en')
out_file.seek(0)
result_html = out_file.read()
```

```python
links = re.findall('"(https?://.*?)"', result_html)
```

```python
links
```

```python
DOWNLOADS_DIR = '/Users/narainyucel/Google Drive/MADS/capstone/legal_nlp/code/test_data'
count = 0
for url in links[:5]: 
    name = url.rsplit('/', 1)[-1]

    filename = os.path.join(DOWNLOADS_DIR, name)

    if not os.path.isfile(filename):
        try:
            urllib.request.urlretrieve(url, filename)
            count += 1
            print(count)
        except HTTPError as error:
            print(error.status, error.reason)
        except URLError as error:
            print(error.reason)
        except TimeoutError:
            print("Request timed out")
        
```
